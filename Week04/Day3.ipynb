```python
# 21. 合并两个有序链表
# https://leetcode.cn/problems/merge-two-sorted-lists/

# Definition for singly-linked list.

class ListNode:
     def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(-1)
        p = dummy
        p1 = list1
        p2 = list2
        while p1 and p2:
            if p1.val < p2.val:
                p.next = p1
                p1 = p1.next
            else:
                p.next = p2
                p2 = p2.next
            p = p.next
        if p1:
            p.next = p1
        if p2:
            p.next = p2
        return dummy.next
```

```python
# 23. 合并 K 个升序链表
# https://leetcode.cn/problems/merge-k-sorted-lists/

# Definition for singly-linked list.

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        import heapq
        dummy = ListNode(-1)
        p = dummy
        head = []
        for i in range(len(lists)):
            while lists[i]:
                heapq.heappush(head, lists[i].val)
                lists[i] = lists[i].next
        while head:
            p.next = ListNode(heapq.heappop(head))
            p = p.next
        return dummy.next
```

```python
# 23. 合并 K 个升序链表
# https://leetcode.cn/problems/merge-k-sorted-lists/

# Definition for singly-linked list.

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        import heapq
        dummy = ListNode(-1)
        p = dummy
        head = []
        for i in range(len(lists)):
            if lists[i]:
                heapq.heappush(head, (lists[i].val, i))
                lists[i] = lists[i].next
        while head:
            val, idx = heapq.heappop(head)
            p.next = ListNode(val)
            p = p.next
            if lists[idx]:
                heapq.heappush(head, (lists[idx].val, idx))
                lists[idx] = lists[idx].next
        return dummy.next
```

```python
'''我们来分析一下代码的逻辑：

创建一个 dummy 节点和一个游标 p, 用于构建合并后的链表。
初始化一个优先队列 pq。
遍历输入的链表列表 lists, 如果当前链表不为空, 则将其头节点加入优先队列 pq。
从优先队列 pq 中取出节点, 节点的值 val 和节点本身 node。
将 node 连接到合并后的链表 p.next, 并将 p 指针后移。
如果 node 还有下一个节点, 将其加入优先队列 pq。
回到步骤 4, 直到优先队列 pq 为空, 即所有节点都被合并到了新链表中。
返回 dummy.next, 即合并后的链表的头节点。'''

'''heapq.heappush(pq, (head.val, head))将链表头节点的值和节点本身作为元组压入优先级队列。
这是因为优先级队列在排序时, 会根据元组的第一个元素进行比较, 也就是head.val的值。
这样做的目的是为了选择当前k个链表中最小的节点进行合并。
由于最小堆的特性, 优先级队列中的节点总是以递增的顺序排列, 因此每次弹出的节点都是当前k个链表中最小的节点。

当第一个元素相同时, heapq 会比较第二个元素, 所以第二个元素不能直接压入 node'''
```

```python
# 23. 合并 K 个升序链表
# https://leetcode.cn/problems/merge-k-sorted-lists/

# Definition for singly-linked list.

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution: # 不使用优先队列
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        lst = []
        for node in lists:
            while node:
                lst.append(node.val)
                node = node.next
        lst.sort()
        dummy = ListNode(-1)
        p = dummy
        # 使用 while lst: v = lst.pop(0) 可能导致性能问题, 虽然也能通过就是了
        for i in range(len(lst)):
            p.next = ListNode(lst[i])
            p = p.next
        return dummy.next
    
'''融会贯通思想才是关键'''
```

```python
# 19. 删除链表的倒数第 N 个结点
# https://leetcode.cn/problems/remove-nth-node-from-end-of-list/

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        m = 0
        dummy = ListNode(-1)
        dummy.next = head
        while head:
            head = head.next
            m += 1
        t = m-n
        p = dummy
        for i in range(t):
            p = p.next
        p.next = p.next.next
        return dummy.next
```

```python
# 19. 删除链表的倒数第 N 个结点
# https://leetcode.cn/problems/remove-nth-node-from-end-of-list/

# 使用一趟扫描实现
# Maintain two pointers and update one with a delay of n steps.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummy = ListNode(-1) # 有可能删除第一个结点，使用 dummy 防止报错
        dummy.next = head
        p1 = dummy
        for _ in range(n): # 因为定义了dummy, 所以 p2 正好是第 n-1 个结点
            p1 = p1.next
        p2 = dummy
        while p1.next:
            p1 = p1.next
            p2 = p2.next
        p2.next = p2.next.next
        return dummy.next
```

```python
# 876. 链表的中间结点
# https://leetcode.cn/problems/middle-of-the-linked-list/

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:
        p1 = head
        p2 = head
        while p1 and p1.next:
            p1 = p1.next.next
            p2 = p2.next
        return p2

[1,2,3,4,5]
[3,4,5]   
[1,2,3,4,5,6]
[4,5,6]
```

```python
# 142. 环形链表 II
# https://leetcode.cn/problems/linked-list-cycle-ii/

# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None


class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        p1 = head
        p2 = head
        while p1 and p1.next:
            p1 = p1.next.next
            p2 = p2.next
            if p1 == p2:
                break
        if not p1 or not p1.next:
            return None
        p2 = head
        while p1 != p2:
            p1 = p1.next
            p2 = p2.next
        return p1
```

```python
# 160. 相交链表
# https://leetcode.cn/problems/intersection-of-two-linked-lists/

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        p1 = headA
        p2 = headB
        while p1 != p2:
            if not p1:
                p1 = headB
            elif not p2:
                p2 = headA
            else:
                p1 = p1.next
                p2 = p2.next
        return p1
```
