```python
# 785. 判断二分图
# https://leetcode.cn/problems/is-graph-bipartite/

# DFS
class Solution:
    def __init__(self) -> None:
        self.ok = True
        self.color = [] # 记录图中节点的颜色，False和True代表两种不同颜色
        self.visited = []

    def isBipartite(self, graph: List[List[int]]) -> bool:
        n = len(graph)
        self.color = [False]*n
        self.visited = [False]*n
        for v in range(n):
            if not self.visited[v]:
                self.traverse(graph, v)
                if not self.ok:
                    break
        return self.ok

    def traverse(self, graph: List[List[int]], v: int) -> None:
        self.visited[v] = True
        for w in graph[v]:
            if not self.visited[w]:
                self.color[w] = not self.color[v]
                self.traverse(graph, w)
            else:
                if self.color[w] == self.color[v]:
                    self.ok = False
                    return
```

```python
# 785. 判断二分图
# https://leetcode.cn/problems/is-graph-bipartite/

# BFS
class Solution:
    def __init__(self) -> None:
        self.ok = True
        self.color = [] # 记录图中节点的颜色，False和True代表两种不同颜色
        self.visited = []

    def isBipartite(self, graph: List[List[int]]) -> bool:
        n = len(graph)
        self.visited = [False]*n
        self.color = [False]*n
        for v in range(n):
            self.BFS(graph, v)
        return self.ok

    def BFS(self, graph, start):
        q = []
        q.append(start)
        self.visited[start] = True
        while q:
            v = q.pop()
            for w in graph[v]:
                if not self.visited[w]:
                    self.color[w] = not self.color[v]
                    self.visited[w] = True
                    q.append(w)
                else:
                    if self.color[w] == self.color[v]:
                        self.ok = False
                        return
```
