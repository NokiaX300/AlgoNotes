```python
# 200. 岛屿数量
# https://leetcode.cn/problems/number-of-islands/

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        def DFS(grid, i, j):
            if not 0<=i<len(grid) or not 0<=j<len(grid[0]) or grid[i][j] != '1':
                return
            grid[i][j] = '*'
            DFS(grid, i+1, j)
            DFS(grid, i-1, j)
            DFS(grid, i, j+1)
            DFS(grid, i, j-1)

        count = 0
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == '1':
                    DFS(grid, i, j)
                    count += 1
        return count
```

```python
# 1254. 统计封闭岛屿的数目
# https://leetcode.cn/problems/number-of-closed-islands/

class Solution:
    def closedIsland(self, grid: List[List[int]]) -> int:
        def DFS(grid, i, j):
            if not 0<=i<len(grid) or not 0<=j<len(grid[0]) or grid[i][j] != 0:
                return
            grid[i][j] = '*'
            DFS(grid, i+1, j)
            DFS(grid, i-1, j)
            DFS(grid, i, j+1)
            DFS(grid, i, j-1)

        for i in range(len(grid)):
            DFS(grid, i, 0)
            DFS(grid, i, len(grid[0])-1)
        for j in range(len(grid[0])):
            DFS(grid, 0, j)
            DFS(grid, len(grid)-1, j)

        res = 0
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 0:
                    res += 1
                    DFS(grid, i, j)
        return res
```

```python
# 1020. 飞地的数量
# https://leetcode.cn/problems/number-of-enclaves/

class Solution:
    def numEnclaves(self, grid: List[List[int]]) -> int:
        m = len(grid)
        n = len(grid[0])
        for i in range(m):
            self.DFS(grid, i, 0)
            self.DFS(grid, i, n-1)
        for j in range(n):
            self.DFS(grid, 0, j)
            self.DFS(grid, m-1, j)

        res = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    res += 1
        return res

    def DFS(self, grid, i, j):
        m = len(grid)
        n = len(grid[0])
        if not 0<=i<m or not 0<=j<n or grid[i][j] != 1:
            return
        grid[i][j] = '*'
        self.DFS(grid, i+1, j)
        self.DFS(grid, i-1, j)
        self.DFS(grid, i, j+1)
        self.DFS(grid, i, j-1)
```

```python
# 547. 省份数量
# https://leetcode.cn/problems/number-of-provinces/

class Solution:
    def findCircleNum(self, isConnected: List[List[int]]) -> int:
        visited = [None]*len(isConnected)
        def DFS(i):
            visited[i] = 1
            for j in range(len(isConnected)):
                if isConnected[i][j] == 1 and not visited[j]:
                    DFS(j)

        count = 0
        for i in range(len(isConnected)):
            if not visited[i]:
                DFS(i)
                count += 1
        return count
```

```python
# 684. 冗余连接
# https://leetcode.cn/problems/redundant-connection/

class Solution:
    def findRedundantConnection(self, edges: List[List[int]]) -> List[int]:
        dic = {}
        def DFS(u, v):
            visited.add(u)
            for ele in dic[u]:
                if ele == v:
                    return True
                elif ele not in visited:
                    if DFS(ele, v):
                        return True
            return False
        
        for edge in edges:
            if edge[0] not in dic:
                dic[edge[0]] = {edge[1]}
            else:
                visited = set() # {} 是创建一个空集合
                if DFS(edge[0], edge[1]):
                    return edge
                else:
                    dic[edge[0]].add(edge[1])
            if edge[1] not in dic:
                dic[edge[1]] = {edge[0]}
            else:
                dic[edge[1]].add(edge[0])
```

```python
# 93. 复原 IP 地址
# https://leetcode.cn/problems/restore-ip-addresses/

class Solution:
    def restoreIpAddresses(self, s: str) -> List[str]:
        if not s or len(s) < 4:
            return []
        res = list()
        self.dfs(s, 0, '', res)
        return res
    
    # DFS递归，s为目标字符串，idx为对s的分割次数（也是递归深度），path为划分的中间（最终）结果字符串，res为结果列表
    def dfs(self, s: str, idx: int, path: str, res: List[str]):
        # 如果idx大于4，那么说明字符串s已经被分割的次数大于4，此时直接返回
        if idx > 4:
            return 
        # 如果idx等于4，那么还要继续判断递归传入的s是否已经为空，如果s为空则说明字符串已经被完全分割为IP地址的形式，此时将该分割方法（路径）存储到结果列表res中，然后返回
        if idx == 4 and not s:
            # 这里符合条件的path的形式应该是"xxx.xxx.xxx.xxx."，因此path[:-1]是为了舍弃最后的'.'字符
            res.append(path[:-1])
            return
        for i in range(len(s)):
            # 后面的if语句用于处理以下两种情况：
            # （1）当s的首字符为'0'时，可以直接将'0'作为IP地址中四个整数之一
            # （2）当s的首字符不为'0'时，需要保证s[:i+1]处于IP地址整数的范围之内
            if s[:i+1] == '0' or (s[0] != '0' and 0 < int(s[:i+1]) < 256):
                # DFS递归调用的参数需要进行以下操作：
                # （1）将下标i之后的s[i+1:]字符串作为新的s进行递归参数传入
                # （2）分割次数idx+1
                # （3）将中间字符串结果path后连接下标i之前的s[:i+1]字符串，并在最后加入'.'
                # （4）结果列表res依然原封不动的传入下次递归
                self.dfs(s[i+1:], idx+1, path+s[:i+1]+'.', res)
```

```python
# 93. 复原 IP 地址
# https://leetcode.cn/problems/restore-ip-addresses/

# 再写一遍
from typing import List
class Solution:
    def restoreIpAddresses(self, s: str) -> List[str]:
        res = []
        self.DFS(s, '', 0, res)
        return res

    def DFS(self, s: str, path: str, idx: int, res: List[str]):
        if idx > 4:
            return
        if idx == 4 and not s:
            res.append(path[:-1])
            return
        for i in range(len(s)):
            if s[:i+1] == '0' or (s[0] != '0' and int(s[:i+1]) < 256):
                self.DFS(s[i+1:], path+s[:i+1]+'.', idx+1, res)
```
