```python
# 力扣 hot100-38
# 543. 二叉树的直径
# https://leetcode.cn/problems/diameter-of-binary-tree/description/?envType=study-plan-v2&envId=top-100-liked

# 后序自底向上
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        self.res = 0
        self.traverse(root)
        return self.res

    def traverse(self, root):
        if not root:
            return 0
        left = self.traverse(root.left)
        right = self.traverse(root.right)
        diameter = left + right
        self.res = max(self.res, diameter)
        # 返回值要该子树的左或右选一条最长的
        return max(left, right)+1
```

```python
# 力扣 hot100-39
# 102. 二叉树的层序遍历
# https://leetcode.cn/problems/binary-tree-level-order-traversal/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        node_list = [root]
        res = []
        while node_list:
            level_list = []
            for _ in range(len(node_list)):
                cur = node_list.pop(0)
                level_list.append(cur.val)
                if cur.left:
                    node_list.append(cur.left)
                if cur.right:
                    node_list.append(cur.right)
            res.append(level_list)
        return res
```

```python
# 力扣 hot100-40
# 108. 将有序数组转换为二叉搜索树
# https://leetcode.cn/problems/convert-sorted-array-to-binary-search-tree/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
        self.nums = nums
        return self.traverse(0, len(nums)-1)

    def traverse(self, left, right):
        if left>right:
            return
        mid = (right-left)//2+left
        root = TreeNode(self.nums[mid])
        root.left = self.traverse(left, mid-1)
        root.right = self.traverse(mid+1, right)
        return root
```

```python
# 力扣 hot100-41
# 98. 验证二叉搜索树
# https://leetcode.cn/problems/validate-binary-search-tree/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        return self.traverse(root, None, None)

    def traverse(self, root, min_root, max_root):
        if not root:
            return True
        if min_root and min_root.val>=root.val:
            return False
        if max_root and max_root.val<=root.val:
            return False
        return self.traverse(root.left, min_root, root) and self.traverse(root.right, root, max_root)
```

```python
# 力扣 hot100-42
# 230. 二叉搜索树中第 K 小的元素
# https://leetcode.cn/problems/kth-smallest-element-in-a-bst/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        self.res = []
        self.traverse(root)
        return self.res[k-1]

    def traverse(self, root):
        if not root:
            return
        self.traverse(root.left)
        self.res.append(root.val)
        self.traverse(root.right)
```

```python
# 力扣 hot100-43
# 199. 二叉树的右视图
# https://leetcode.cn/problems/binary-tree-right-side-view/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        node_list = [root]
        res = []
        while node_list:
            level_list = []
            for _ in range(len(node_list)):
                cur = node_list.pop(0)
                level_list.append(cur.val)
                if cur.left:
                    node_list.append(cur.left)
                if cur.right:
                    node_list.append(cur.right)
            res.append(level_list[-1])
        return res
```

```python
# 力扣 hot100-44
# 114. 二叉树展开为链表
# https://leetcode.cn/problems/flatten-binary-tree-to-linked-list/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def flatten(self, root: Optional[TreeNode]) -> None:
        if not root:
            return
        self.flatten(root.left)
        self.flatten(root.right)
        # 想象左右子树都已拉伸完成
        left_tree = root.left
        right_tree = root.right
        root.left = None
        root.right = left_tree
        p = root
        while p.right:
            p = p.right
        p.right = right_tree
```

```python
# 力扣 hot100-45
# 105. 从前序与中序遍历序列构造二叉树
# https://leetcode.cn/problems/construct-binary-tree-from-preorder-and-inorder-traversal/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        self.var_to_ind = dict()
        self.pre = preorder
        self.ino = inorder
        for i,c in enumerate(inorder):
            self.var_to_ind[c] = i
        return self.traverse(0, len(preorder)-1, 0, len(inorder)-1)
    
    def traverse(self, pre_start, pre_end, in_start, in_end):
        if pre_end<pre_start:
            return
        root = TreeNode(self.pre[pre_start])
        ind = self.var_to_ind[root.val]
        root.left = self.traverse(pre_start+1, pre_start+ind-in_start, in_start, ind-1)
        root.right = self.traverse(pre_start+ind-in_start+1, pre_end, ind+1, in_end)
        return root
```

```python
# 力扣 hot100-46
# 437. 路径总和 III
# https://leetcode.cn/problems/path-sum-iii/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
        if not root:
            return 0
        # 从根出发的
        self.res = self.traverse(root, targetSum)
        # 加上从子树出发的
        self.res += self.pathSum(root.left, targetSum)
        self.res += self.pathSum(root.right, targetSum)
        return self.res

    # 自顶向下
    def traverse(self, root, target):
        if not root:
            return 0
        summ = 0
        if root.val == target:
            summ += 1
        summ += self.traverse(root.left, target-root.val)
        summ += self.traverse(root.right, target-root.val)
        return summ
```
