```python
# 力扣 hot100-40
# 543. 二叉树的直径
# https://leetcode.cn/problems/diameter-of-binary-tree/description/?envType=study-plan-v2&envId=top-100-liked

# 后序自底向上
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        self.res = 0
        self.traverse(root)
        return self.res

    def traverse(self, root):
        if not root:
            return 0
        left = self.traverse(root.left)
        right = self.traverse(root.right)
        diameter = left + right
        self.res = max(self.res, diameter)
        # 返回值要该子树的左或右选一条最长的
        return max(left, right)+1
```

```python
# 力扣 hot100-41
# 102. 二叉树的层序遍历
# https://leetcode.cn/problems/binary-tree-level-order-traversal/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        node_list = [root]
        res = []
        while node_list:
            level_list = []
            for _ in range(len(node_list)):
                cur = node_list.pop(0)
                level_list.append(cur.val)
                if cur.left:
                    node_list.append(cur.left)
                if cur.right:
                    node_list.append(cur.right)
            res.append(level_list)
        return res
```

```python
# 力扣 hot100-42
# 108. 将有序数组转换为二叉搜索树
# https://leetcode.cn/problems/convert-sorted-array-to-binary-search-tree/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
        self.nums = nums
        return self.traverse(0, len(nums)-1)

    def traverse(self, left, right):
        if left>right:
            return
        mid = (right-left)//2+left
        root = TreeNode(self.nums[mid])
        root.left = self.traverse(left, mid-1)
        root.right = self.traverse(mid+1, right)
        return root
```

```python
# 力扣 hot100-43
# 98. 验证二叉搜索树
# https://leetcode.cn/problems/validate-binary-search-tree/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        return self.traverse(root, None, None)

    def traverse(self, root, min_root, max_root):
        if not root:
            return True
        if min_root and min_root.val>=root.val:
            return False
        if max_root and max_root.val<=root.val:
            return False
        return self.traverse(root.left, min_root, root) and self.traverse(root.right, root, max_root)
```

```python
# 力扣 hot100-47
# 105. 从前序与中序遍历序列构造二叉树
# https://leetcode.cn/problems/construct-binary-tree-from-preorder-and-inorder-traversal/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        self.var_to_ind = dict()
        self.pre = preorder
        self.ino = inorder
        for i,c in enumerate(inorder):
            self.var_to_ind[c] = i
        return self.traverse(0, len(preorder)-1, 0, len(inorder)-1)
    
    def traverse(self, pre_start, pre_end, in_start, in_end):
        if pre_end<pre_start:
            return
        root = TreeNode(self.pre[pre_start])
        ind = self.var_to_ind[root.val]
        root.left = self.traverse(pre_start+1, pre_start+ind-in_start, in_start, ind-1)
        root.right = self.traverse(pre_start+ind-in_start+1, pre_end, ind+1, in_end)
        return root
```

```python
# 力扣 hot100-48
# 437. 路径总和 III
# https://leetcode.cn/problems/path-sum-iii/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
        if not root:
            return 0
        # 从根出发的
        self.res = self.traverse(root, targetSum)
        # 加上从子树出发的
        self.res += self.pathSum(root.left, targetSum)
        self.res += self.pathSum(root.right, targetSum)
        return self.res

    # 自顶向下
    def traverse(self, root, target):
        if not root:
            return 0
        summ = 0
        if root.val == target:
            summ += 1
        summ += self.traverse(root.left, target-root.val)
        summ += self.traverse(root.right, target-root.val)
        return summ
```

```python
# 力扣 hot100-49
# 236. 二叉树的最近公共祖先
# https://leetcode.cn/problems/lowest-common-ancestor-of-a-binary-tree/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        # 边界条件是叶子或者找到 p 或 q
        # 注意是自底向上, 最后返回的 p 或 q 就是 lca
        if root in (p, q, None):
            return root
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)
        # 一个在左子树一个在右子树
        if left and right:
            return root
        # 都不在左子树说明 p q 都在右子树, 因而 lca 在右子树
        if not left:
            return right
        return left
```

```python
# 力扣 hot100-50
# 124. 二叉树中的最大路径和
# https://leetcode.cn/problems/binary-tree-maximum-path-sum/description/?envType=study-plan-v2&envId=top-100-liked

# 代码参考 543. 二叉树的直径
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        self.res = -float('inf')
        self.traverse(root)
        return self.res

    def traverse(self, root):
        if not root:
            return 0
        left = self.traverse(root.left)
        right = self.traverse(root.right)
        # 直径最大值
        self.res = max(self.res, left+right+root.val)
        # 如果某一侧的路径和为负数, 舍弃该路径
        # 注意返回值是选一条子树
        return max(max(left, right)+root.val, 0)
```

```python
# 力扣 hot100-51
# 200. 岛屿数量
# https://leetcode.cn/problems/number-of-islands/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        self.grid = grid
        self.m = len(grid)
        self.n = len(grid[0])
        self.res = 0
        for i in range(self.m):
            for j in range(self.n):
                if self.grid[i][j]=='1':
                    self.res += 1
                    self.dfs(i, j)
        return self.res

    def dfs(self, i, j):
        if not 0<=i<self.m or not 0<=j<self.n or not self.grid[i][j]=='1':
            return
        self.grid[i][j] = '%'
        self.dfs(i+1, j)
        self.dfs(i-1, j)
        self.dfs(i, j+1)
        self.dfs(i, j-1)
```

```python
# 力扣 hot100-52
# 994. 腐烂的橘子
# https://leetcode.cn/problems/rotting-oranges/description/?envType=study-plan-v2&envId=top-100-liked

# 首先要意识到是 BFS
# 套用层序遍历模板
class Solution:
    def orangesRotting(self, grid: List[List[int]]) -> int:
        m = len(grid)
        n = len(grid[0])
        node_list = []
        fresh = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 2:
                    node_list.append((i, j))
                elif grid[i][j] == 1:
                    fresh += 1
        res = 0
        while node_list:
            for _ in range(len(node_list)):
                x, y = node_list.pop(0)
                # 学习新写法
                for i, j in (x-1, y), (x+1, y), (x, y-1), (x, y+1):
                    if 0<=i<m and 0<=j<n and grid[i][j]==1:
                        fresh -= 1
                        grid[i][j] = 2
                        node_list.append((i, j))
            res += 1
        # 还剩新鲜橘子
        if fresh:
            return -1
        # 最后一次加入腐烂橘子后遍历不再增加时间, 所以 res-1
        # 0 是为了应对一个橘子都没有的情况
        return max(0, res-1)
```

```python
# 力扣 hot100-53
# 207. 课程表
# https://leetcode.cn/problems/course-schedule/?envType=study-plan-v2&envId=top-100-liked

# BFS
from collections import defaultdict
class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        pre = defaultdict(list)
        post = defaultdict(list)
        for i, j in prerequisites:
            pre[i].append(j)
            post[j].append(i)
        indegree = []
        unlearn = []
        learned = []
        for i in range(numCourses):
            indegree.append(len(pre[i]))
        for i in range(len(indegree)):
            if indegree[i]==0:
                unlearn.append(i)
        while unlearn:
            cur = unlearn.pop(0)
            learned.append(cur)
            for i in post[cur]:
                indegree[i] -= 1
                if indegree[i] == 0:
                    unlearn.append(i)
        return len(learned)==numCourses
```

```python
# 力扣 hot100-54
# 208. 实现 Trie (前缀树)
# https://leetcode.cn/problems/implement-trie-prefix-tree/description/?envType=study-plan-v2&envId=top-100-liked

class Node:
    def __init__(self) -> None:
        # 子节点列表
        self.children = [None]*26
        # 标记是否尾节点
        self.isEnd = False


class Trie:
    def __init__(self):
        self.root = Node()

    def insert(self, word: str) -> None:
        node = self.root
        for char in word:
            id = ord(char)-ord('a')
            if not node.children[id]:
                node.children[id] = Node()
            node = node.children[id]
        node.isEnd = True

    def search(self, word: str) -> bool:
        node = self.searchPrefix(word)
        return node!=None and node.isEnd

    def startsWith(self, prefix: str) -> bool:
        return self.searchPrefix(prefix)!=None

    def searchPrefix(self, word):
        node = self.root
        for char in word:
            id = ord(char)-ord('a')
            node = node.children[id]
            if not node:
                return
        return node
```

```python
# 力扣 hot100-55
# 46. 全排列
# https://leetcode.cn/problems/permutations/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        self.track = []
        self.res = []
        self.used = [False]*len(nums)
        self.backtrack(nums)
        return self.res

    def backtrack(self, nums):
        if len(self.track)==len(nums):
            self.res.append(self.track.copy())
        for i in range(len(nums)):
            if self.used[i]:
                continue
            self.used[i] = True
            self.track.append(nums[i])
            self.backtrack(nums)
            self.track.pop()
            self.used[i] = False
```

```python
# 力扣 hot100-57
# 17. 电话号码的字母组合
# https://leetcode.cn/problems/letter-combinations-of-a-phone-number/?envType=study-plan-v2&envId=top-100-liked

# for →
# start ↓
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []
        self.map = ['', '', "abc", "def", "ghi", "jkl", "mno", "pqrs", "tuv", "wxyz"]
        self.track = []
        self.res = []
        self.backtrack(digits, 0)
        return self.res

    def backtrack(self, digits, start):
        if len(self.track)==len(digits):
            self.res.append(''.join(self.track.copy()))
            return
        digit = ord(digits[start])-ord('0')
        for c in self.map[digit]:
            self.track.append(c)
            self.backtrack(digits, start+1)
            self.track.pop()
```

```python
# 力扣 hot100-58
# 39. 组合总和
# https://leetcode.cn/problems/combination-sum/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        self.track = []
        self.res = []
        self.summ = 0
        self.backtrack(candidates, 0, target)
        return self.res

    def backtrack(self, nums, start, target):
        if self.summ == target:
            self.res.append(self.track.copy())
        if self.summ > target:
            return
        for i in range(start, len(nums)):
            self.summ += nums[i]
            self.track.append(nums[i])
            self.backtrack(nums, i, target)
            self.track.pop()
            self.summ -= nums[i]
```

```python
# 力扣 hot100-59
# 22. 括号生成
# https://leetcode.cn/problems/generate-parentheses/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        self.track = []
        self.res = []
        self.backtrack(n, n)
        return self.res

    def backtrack(self, left, right):
        if left>right:
            return
        if left<0 or right<0:
            return
        if left == 0 and right == 0:
            self.res.append(''.join(self.track))
            return
        self.track.append('(')
        self.backtrack(left-1, right)
        self.track.pop()
        self.track.append(')')
        self.backtrack(left, right-1)
        self.track.pop()
```

```python
# 力扣 hot100-60
# 79. 单词搜索
# https://leetcode.cn/problems/word-search/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        self.m = len(board)
        self.n = len(board[0])
        for i in range(self.m):
            for j in range(self.n):
                if self.dfs(board, word, i, j, 0):
                    return True
        return False

    def dfs(self, board, word, i, j, k):
        if board[i][j] != word[k]:
            return False
        if k==len(word)-1:
            return True
        board[i][j] = 0
        for x,y in (i,j-1), (i,j+1), (i-1,j), (i+1,j):
            if 0<=x<self.m and 0<=y<self.n and self.dfs(board, word, x, y, k+1):
                return True
        board[i][j] = word[k]
        return False
```

```python
# 力扣 hot100-61
# 131. 分割回文串
# https://leetcode.cn/problems/palindrome-partitioning/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def partition(self, s: str) -> List[List[str]]:
        n = len(s)
        res = []
        path = []
        def dfs(i):
            if i==n:
                res.append(path.copy())
            for j in range(i, n):
                t = s[i:j+1]
                if t==t[::-1]:
                    path.append(t)
                    dfs(j+1)
                    path.pop()
        dfs(0)
        return res
```

```python
# 力扣 hot100-62
# 51. N 皇后
# https://leetcode.cn/problems/n-queens/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        self.res = []
        board = [['.']*n for _ in range(n)]
        self.backtrack(board, 0)
        return self.res

    def backtrack(self, board: List[List[str]], row: int) -> None:
        if row == len(board):
            # self.res.append([row[:] for row in board]) 只复制了列表的引用，而不是列表的内容
            self.res.append([''.join(row) for row in board])
            return

        for j in range(len(board[row])):
            if not self.isValid(board, row, j):
                continue
            board[row][j] = 'Q'
            self.backtrack(board, row+1)
            board[row][j] = '.'

    # 只需检查左上角，右上角和上方的格子
    def isValid(self, board: List[str], row: int, col: int) -> bool:
        n = len(board)
        for i in range(n):
            if board[i][col] == 'Q':
                return False
        r = row-1
        c = col-1
        while r>=0 and c>=0:
            if board[r][c] == 'Q':
                return False
            r -= 1
            c -= 1
        r = row-1
        c = col+1
        while r>=0 and c<=n-1:
            if board[r][c] == 'Q':
                return False
            r -= 1
            c += 1
        return True
```

```python
# 力扣 hot100-65
# 34. 在排序数组中查找元素的第一个和最后一个位置
# https://leetcode.cn/problems/find-first-and-last-position-of-element-in-sorted-array/description/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        if target not in nums:
            return [-1,-1]
        l = self.find_left(nums, target)
        r = self.find_left(nums, target+1)
        return [l, r-1]

    def find_left(self, nums, target):
        left = 0
        right = len(nums)-1
        while left<=right:
            mid = (right-left)//2+left
            if nums[mid]<target:
                left = mid+1
            else:
                right = mid-1
        return left
```

```python
# 力扣 hot100-66
# 33. 搜索旋转排序数组
# https://leetcode.cn/problems/search-in-rotated-sorted-array/?envType=study-plan-v2&envId=top-100-liked

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left = 0
        right = len(nums)-1
        while left<=right:
            mid = left+(right-left)//2
            if nums[mid] == target:
                return mid
            if nums[left] <= nums[mid]:
                if nums[left] <= target <= nums[mid]:
                    right = mid-1
                else:
                    left = mid+1
            else:
                if nums[mid] <= target <= nums[right]:
                    left = mid+1
                else:
                    right = mid-1
        return -1
```
